```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()

#NOTE : tip -> dependent feature  ,  sex, smoker , time -> independent Feature
```

## correlation with heat map

A correlation heatmap uses colored cells, typically in a monocromatic scale, to show a 2D correlation matrix(table) between two discrite dimensions or event types. it is very important in feature selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joint Plot

A join plot allows to study the relationship between 2 numeric variables . The central chart display their correlation. it is usually a scatterplot, a hexbin plot, a 2D histogram or a 2Ddensity Plot

## univariate analysis

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind='hex')
```

```python
sns.jointplot(x = 'tip',y = 'size',data = df,kind='hex')
```

```python
sns.jointplot(x = 'tip',y = 'total_bill',data = df,kind='reg')  # reg=> regression 
```

## Pair Plot

A "pairplot" is also known as a scatterplot, in which one variable in the same data row is matched with another variables value, like this. Pairs plots are just elaborations on this , showing all variables paired with all the other variables.

```python
#NOTE : for Pairplot it must be integer or float values
sns.pairplot(df)

# it is permutation  and Combination of all the values
```

```python
sns.pairplot(df,hue='sex') 

# we use hue parameter for getting result according to 'sex'
```

```python
sns.pairplot(df,hue='smoker') 
```

```python
df['smoker'].value_counts()
```

## Distplot
-  dist plot help us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde = False, bins = 10)
```

# Categorical Plots

seaborn also helps us in doing the analysis on categorical data points.
- boxplot
- violin plot
- countplot
- bar plot

## CountPlot

```python
# Count Plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y = 'sex',data=df)
```

## Bar Plot

In barplot we give both x and y values 

```python
sns.barplot(x = 'total_bill', y = 'sex', data = df)
```

```python
sns.barplot(x = 'total_bill', y = 'smoker', data = df)
```

## Box Plot

A box and whiskerplot is graph tht presents information from a five number summary

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df, palette ="rainbow")  
#  palette ="rainbow"   =>  we using it for giving colour 
```

```python
sns.boxplot(data =df, orient = 'v')

## boxplot gives us result on whole dataset 
```

```python
# categorize my data based on same other categories
sns.boxplot(x = 'total_bill', y = 'day',hue = "smoker", data =df)

# hue is basically shows how the model is differnt than other, who have max outliers 
```

## Violin Plot

violin plot helps us to see both the distribution of data in terms of Kernal Density Estimation and the box Plot

```python
sns.violinplot(x = 'total_bill',y = 'day',data = df , palette='rainbow')
```
