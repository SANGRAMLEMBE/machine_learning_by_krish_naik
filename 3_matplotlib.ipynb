# plot


```python
# compute the x and y coordinates for points on a sin Curve

x = np.arange(0,4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

# Subplot

```python
# subplot()
# compute the x and y coordinate for points on sine and cosine curves
x = np.arange(0, 5*np.pi , 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a subplot grid that has height 2 and width 1, 2 rows 1 col
# and set the first such subplot as active
plt.subplot(2,1,1)

# make the first plot
plt.plot(x,y_sin,'r--')
plt.title('sine')

# set the second subplot as active, and make the secound plot

plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cosine')

# show the fig
plt.show()
```

# bar plot

```python
## bar plot
x = [2,8,10,3,5,1,8,5,12]
y = [11,16,9,8,20,14,16,19,13]

x2 = [3,9,11,4,5,9,6.5]
y2 = [6,15,7,20,14,9,17]

plt.bar(x,y,align = 'center')
plt.bar(x2 ,y2 , color ='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

# Pie chart

```python
# data to plot
labels = 'python', 'C++', 'Ruby','Java'
sizes = [315,130,245,210]
colors =['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.1,0,0,0.2) #explode 1st slice

#Plot
plt.pie(sizes , explode = explode , labels = labels , colors = colors,autopct = '%1.1f%%',shadow = True)

plt.axis('equal')
plt.show()

# explode: it is used to make gap betn the diff ccharacteristic 
```
